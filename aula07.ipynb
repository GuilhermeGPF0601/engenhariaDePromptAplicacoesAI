# 1. Soma de Listas

```python
# Define uma lista de números como exemplo
numeros = [10, 20, 30, 40, 50]

# Calcula a soma dos elementos na lista usando a função sum()
soma = sum(numeros)

# Exibe a soma
print(f"A soma dos números é: {soma}")
```

# 2. Número Primo

```python
def eh_primo(n):
    # Números menores que 2 não são primos por definição.
    if n < 2:
        return False
    # Itera de 2 até n-1. Se 'n' for divisível por qualquer número nesse intervalo, não é primo.
    for i in range(2, n):
        if n % i == 0:
            return False
    # Se nenhum divisor for encontrado, o número é primo.
    return True

def verifica_e_imprime_primo(numero):
    """Verifica se um número é primo e imprime o resultado."""
    if eh_primo(numero):
        print(f"{numero} é um número primo.")
    else:
        print(f"{numero} não é um número primo.")

# Exemplos de uso:
verifica_e_imprime_primo(17)
verifica_e_imprime_primo(10)
```

# 3. Inversão de String

```python
def inverte_texto(texto):
    # Utiliza fatiamento de string (slicing) para inverter o texto de forma concisa.
    return texto[::-1]

# Exemplos de testes:
print(f"'Olá mundo' invertido é: {inverte_texto('Olá mundo')}")
print(f"'Python' invertido é: {inverte_texto('Python')}")
print(f"'12345' invertido é: {inverte_texto('12345')}")
```
