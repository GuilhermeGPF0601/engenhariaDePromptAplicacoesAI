```python
# Define uma função para verificar a temperatura
def verificar_temperatura(temperatura):
    # Verifica se a temperatura é maior que 30 graus Celsius
    if temperatura > 30:
        # Se for, exibe um alerta de superaquecimento usando concatenação de string
        print("ALERTA: Superaquecimento detectado! Temperatura atual: " + str(temperatura) + "°C")
    else:
        # Caso contrário, exibe que a temperatura está normal usando concatenação de string
        print("Temperatura normal. Temperatura atual: " + str(temperatura) + "°C")

# --- Exemplos de uso da função ---

# Simula uma temperatura atual de 35°C (situação de superaquecimento)
temperatura_atual = 35
verificar_temperatura(temperatura_atual)

# Simula uma temperatura normal de 25°C
temperatura_normal = 25
verificar_temperatura(temperatura_normal)
```

# 1. Organização de Pastas

```python
import os
import shutil
import random
import string

def organizar_arquivos_por_extensao(diretorio_origem):
    if not os.path.isdir(diretorio_origem):
        print("Erro: O diretório '" + diretorio_origem + "' não existe.")
        return

    print("\nOrganizando arquivos em: " + diretorio_origem)

    for nome_arquivo in os.listdir(diretorio_origem):
        caminho_completo_arquivo = os.path.join(diretorio_origem, nome_arquivo)

        # Ignora diretórios e links simbólicos
        if os.path.isfile(caminho_completo_arquivo):
            nome_base, extensao = os.path.splitext(nome_arquivo)
            extensao = extensao.lower() # Garante que a extensão seja minúscula para consistência

            # Se não houver extensão (ex: 'Makefile'), move para uma pasta 'Sem Extensao'
            if not extensao:
                pasta_destino = os.path.join(diretorio_origem, "Sem Extensao")
            else:
                # Remove o ponto da extensão para usar como nome da pasta (ex: '.pdf' -> 'pdf')
                pasta_destino = os.path.join(diretorio_origem, extensao[1:])

            # Cria a pasta de destino se ela não existir
            if not os.path.exists(pasta_destino):
                os.makedirs(pasta_destino)
                print("Pasta '" + pasta_destino + "' criada.")

            # Move o arquivo para a pasta de destino
            try:
                shutil.move(caminho_completo_arquivo, pasta_destino)
                print("Movido: '" + nome_arquivo + "' para '" + pasta_destino + "'")
            except PermissionError:
                print("Erro de permissão ao mover '" + nome_arquivo + "'. Verifique as permissões do diretório de origem e destino.")
            except Exception as e:
                print("Erro inesperado ao mover '" + nome_arquivo + "': " + str(e))

    print("Organização concluída em '" + diretorio_origem + "'!")

# --- Exemplos de uso com 10 testes diferentes ---

def gerar_nome_aleatorio(length=8):
    letters = string.ascii_lowercase
    return ''.join(random.choice(letters) for i in range(length))

for i in range(1, 11):
    test_dir_name = "/content/teste_organizacao_" + str(i)
    os.makedirs(test_dir_name, exist_ok=True)
    print("\n--- Executando Teste " + str(i) + ": " + test_dir_name + " ---")

    # Gerar alguns arquivos de teste
    extensoes_comuns = {
        "documento": [".pdf", ".docx", ".txt"],
        "imagem": [".jpg", ".png", ".gif"],
        "planilha": [".xlsx", ".csv"],
        "codigo": [".py", ".js", ".html"],
        "audio": [".mp3", ".wav"]
    }

    # Crie arquivos para cada tipo de extensão
    for tipo, extensoes in extensoes_comuns.items():
        for ext in extensoes:
            file_name = gerar_nome_aleatorio() + ext
            with open(os.path.join(test_dir_name, file_name), "w") as f: f.write("conteúdo de teste")

    # Arquivo com extensão em maiúsculas
    with open(os.path.join(test_dir_name, gerar_nome_aleatorio() + ".JPG"), "w") as f: f.write("conteúdo de teste")
    with open(os.path.join(test_dir_name, gerar_nome_aleatorio() + ".PDF"), "w") as f: f.write("conteúdo de teste")

    # Arquivo sem extensão
    with open(os.path.join(test_dir_name, gerar_nome_aleatorio()), "w") as f: f.write("conteúdo de teste")

    # Arquivo com múltiplos pontos no nome
    with open(os.path.join(test_dir_name, "meu.arquivo.versao1.0.txt"), "w") as f: f.write("conteúdo de teste")

    # Arquivo que já está no tipo de pasta correta (não deve ser movido)
    os.makedirs(os.path.join(test_dir_name, "pdf"), exist_ok=True)
    with open(os.path.join(test_dir_name, "pdf", "ja_organizado.pdf"), "w") as f: f.write("conteúdo de teste")

    # Criar um subdiretório (deve ser ignorado)
    os.makedirs(os.path.join(test_dir_name, "sub_pasta"), exist_ok=True)
    with open(os.path.join(test_dir_name, "sub_pasta", "arquivo_na_subpasta.txt"), "w") as f: f.write("conteúdo de teste")

    organizar_arquivos_por_extensao(test_dir_name)

    # Limpeza: Remover a pasta de teste e todo o seu conteúdo
    try:
        shutil.rmtree(test_dir_name)
        print("Pasta de teste '" + test_dir_name + "' e conteúdo removidos com sucesso.")
    except OSError as e:
        print("Erro ao remover a pasta de teste '" + test_dir_name + "': " + e.strerror)

print("\nTodos os 10 testes foram executados!")
```

# 2. Consulta CEP

```python
import requests # Importa a biblioteca requests para fazer requisições HTTP a APIs externas.

def consultar_cep(cep):
    """
    Consulta a API pública ViaCEP para obter informações detalhadas de endereço a partir de um CEP.

    Args:
        cep (str): O Código de Endereçamento Postal a ser consultado. Pode conter formatação (ex: '01001-000').

    Returns:
        dict or None: Um dicionário contendo os dados do endereço se a consulta for bem-sucedida,
                      ou None se ocorrer algum erro (CEP inválido, não encontrado, problema de conexão, etc.).
    """
    # 1. Pré-processamento do CEP:
    # Remove quaisquer caracteres não numéricos (hífens, pontos, espaços) para garantir que apenas dígitos permaneçam.
    # Isso padroniza o CEP para o formato esperado pela API (8 dígitos numéricos).
    cep = str(cep).replace("-", "").replace(".", "").replace(" ", "")

    # 2. Validação básica do CEP:
    # Verifica se o CEP, após o pré-processamento, tem exatamente 8 dígitos e se todos são numéricos.
    # Se não atender a esses critérios, é considerado um CEP inválido e uma mensagem de erro é exibida.
    if len(cep) != 8 or not cep.isdigit():
        print("Erro: O CEP deve conter 8 dígitos numéricos válidos.")
        return None

    # 3. Construção da URL da API:
    # Monta a URL completa para a requisição à API ViaCEP, inserindo o CEP formatado.
    url = f"https://viacep.com.br/ws/{cep}/json/"

    # 4. Requisição à API e tratamento de erros:
    try:
        # Faz uma requisição GET para a URL da API.
        response = requests.get(url)
        # Verifica se a requisição foi bem-sucedida (status code 2xx).
        # Se for um status de erro (4xx ou 5xx), levanta um HTTPError.
        response.raise_for_status()

        # Tenta decodificar a resposta JSON da API.
        data = response.json()

        # 5. Verificação de erro na resposta da API:
        # A API ViaCEP retorna um campo 'erro': true se o CEP não for encontrado.
        if "erro" in data and data["erro"] == True:
            print(f"Erro: CEP '{cep}' não encontrado na base de dados do ViaCEP.")
            return None
        else:
            # Se não houver erro, retorna o dicionário com os dados do endereço.
            return data

    # 6. Tratamento de exceções específicas da biblioteca requests:
    except requests.exceptions.HTTPError as e:
        # Erro HTTP (ex: 404 Not Found, 500 Internal Server Error).
        print(f"Erro HTTP ao consultar o CEP {cep}: {e}. Verifique o formato do CEP e tente novamente.")
    except requests.exceptions.ConnectionError as e:
        # Erro de conexão (ex: sem internet, servidor ViaCEP fora do ar).
        print(f"Erro de conexão ao consultar o CEP {cep}: {e}. Verifique sua conexão com a internet ou tente mais tarde.")
    except requests.exceptions.Timeout as e:
        # Tempo limite excedido para a requisição.
        print(f"Tempo limite excedido ao consultar o CEP {cep}: {e}. A API demorou muito para responder.")
    except requests.exceptions.RequestException as e:
        # Captura qualquer outro erro inesperado da biblioteca requests.
        print(f"Ocorreu um erro inesperado na requisição para o CEP {cep}: {e}.")
    except ValueError:
        # Erro ao decodificar JSON (resposta da API não é um JSON válido).
        print(f"Erro: Resposta inválida da API para o CEP {cep}. Não foi possível decodificar o JSON.")

    # Em caso de qualquer erro tratado nas exceções, retorna None.
    return None

def exibir_endereco_formatado(dados_endereco):
    """
    Exibe as informações de endereço de forma legível e formatada.

    Args:
        dados_endereco (dict or None): Um dicionário contendo os dados do endereço,
                                       conforme retornado pela função `consultar_cep`.
    """
    # Verifica se os dados do endereço foram fornecidos (não são None).
    if dados_endereco:
        print("\n--- Endereço Encontrado ---") # Cabeçalho para a saída.
        # Usa .get() para acessar os valores do dicionário. Isso evita erros se uma chave não existir.
        print(f"CEP: {dados_endereco.get('cep', 'N/A')}")
        print(f"Logradouro: {dados_endereco.get('logradouro', 'N/A')}")
        print(f"Complemento: {dados_endereco.get('complemento', 'N/A')}")
        print(f"Bairro: {dados_endereco.get('bairro', 'N/A')}")
        print(f"Localidade: {dados_endereco.get('localidade', 'N/A')}")
        print(f"UF: {dados_endereco.get('uf', 'N/A')}")
        print(f"DDD: {dados_endereco.get('ddd', 'N/A')}")
        print("--------------------------") # Rodapé para a saída.
    else:
        # Mensagem exibida se não foi possível obter os dados do endereço.
        print("Não foi possível obter os dados do endereço para o CEP informado.")

# --- Uso interativo do script ---
# Este loop permite que o usuário digite múltiplos CEPs até que decida sair.
while True:
    # Solicita ao usuário que digite um CEP ou 'sair' para encerrar.
    cep_digitado = input("Digite o CEP para consulta (ou 'sair' para encerrar): ")

    # Converte a entrada para minúsculas e verifica se é 'sair'.
    if cep_digitado.lower() == 'sair':
        print("Consulta encerrada. Obrigado por usar o consultor de CEPs!")
        break # Sai do loop, encerrando o programa.

    # Chama a função para consultar o CEP digitado.
    dados = consultar_cep(cep_digitado)
    # Chama a função para exibir o endereço formatado com base nos dados obtidos.
    exibir_endereco_formatado(dados)
```

# 3. Sistema de Notificações

```python
import time # Importa o módulo time para usar a função sleep (pausar a execução).
import random # Importa o módulo random para gerar números aleatórios.

# --- Funções do Sistema de Notificações ---

def gerar_temperatura_simulada():
    """
    Simula a leitura de uma temperatura aleatória entre 20.0°C e 40.0°C.
    Esta função representa um sensor lendo dados do ambiente.

    Returns:
        float: Um valor de temperatura simulado.
    """
    return random.uniform(20.0, 40.0)

def verificar_condicao_critica(temperatura):
    """
    Verifica se a temperatura atual atingiu ou excedeu um limite crítico pré-definido.

    Args:
        temperatura (float): A temperatura atual a ser avaliada.

    Returns:
        bool: True se a temperatura for maior que o limite crítico (30.0°C), indicando
              uma condição crítica (superaquecimento). Caso contrário, retorna False.
    """
    limite_critico = 30.0 # Define o valor de temperatura que é considerado crítico.
    if temperatura > limite_critico:
        return True # Condição crítica detectada.
    else:
        return False # Temperatura dentro dos limites normais.

def enviar_alerta(mensagem):
    """
    Simula o envio de uma mensagem de alerta. Em um sistema real, isso poderia
    envolver o envio de e-mails, SMS, notificações push, ou acionamento de alarmes.
    Neste exemplo, o alerta é exibido no console.

    Args:
        mensagem (str): A string contendo a mensagem de alerta a ser exibida/enviada.
    """
    print("\n--- ALERTA CRÍTICO ---") # Cabeçalho visual para o alerta.
    print("MENSAGEM: " + mensagem) # Exibe a mensagem de alerta.
    print("----------------------") # Rodapé visual para o alerta.

# --- Programa Autônomo de Monitoramento ---

def monitorar_sistema_autonomo(intervalo_segundos=5):
    """
    Inicia um ciclo contínuo de monitoramento autônomo do sistema.
    A cada 'intervalo_segundos', a temperatura é verificada e, se crítica,
    um alerta é disparado.

    Args:
        intervalo_segundos (int): O tempo de espera (em segundos) entre uma verificação
                                  e a próxima. Ajuda a controlar a frequência do monitoramento.
    """
    print("Iniciando monitoramento autônomo... (Pressione 'Stop' ou interrupção do kernel para parar)")
    try:
        while True: # Loop infinito para garantir o monitoramento contínuo.
            temperatura_atual = gerar_temperatura_simulada() # Obtém uma nova leitura de temperatura.
            print("Verificando temperatura... Temperatura atual: " + str(round(temperatura_atual, 2)) + "°C")

            # Verifica se a temperatura atual ultrapassa o limite crítico.
            if verificar_condicao_critica(temperatura_atual):
                # Constrói a mensagem de alerta com a temperatura atual formatada.
                mensagem_alerta = "Superaquecimento detectado! Temperatura: " + str(round(temperatura_atual, 2)) + "°C. Ação necessária!"
                enviar_alerta(mensagem_alerta) # Envia o alerta.
            else:
                print("Temperatura normal.") # Informa que a temperatura está dentro do esperado.

            time.sleep(intervalo_segundos) # Pausa a execução pelo tempo definido antes da próxima iteração.

    except KeyboardInterrupt:
        # Este bloco é executado se o usuário interromper o script manualmente (ex: Ctrl+C ou botão 'Stop').
        print("\nMonitoramento autônomo interrompido pelo usuário.")
    except Exception as e:
        # Captura qualquer outro erro inesperado que possa ocorrer durante o monitoramento.
        print("Erro inesperado no monitoramento: " + str(e))

# Inicia o processo de monitoramento autônomo com um intervalo de 5 segundos.
# Este é o ponto de entrada principal do script.
monitorar_sistema_autonomo(intervalo_segundos=5)
```
