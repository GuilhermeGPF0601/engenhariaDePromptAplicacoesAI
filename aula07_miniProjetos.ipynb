# Calculadora Simples

```python
# Esta função realiza a operação de adição entre dois números.
def adicao(a, b):
    return a + b

# Esta função realiza a operação de subtração entre dois números.
def subtracao(a, b):
    return a - b

# Esta função realiza a operação de multiplicação entre dois números.
def multiplicacao(a, b):
    return a * b

# Esta função realiza a operação de divisão entre dois números.
# Ela verifica se o divisor é zero para evitar um erro de divisão por zero.
def divisao(a, b):
    if b != 0:
        # Converte 'a' para float para garantir que a divisão retorne um número decimal.
        return float(a) / b
    else:
        # Retorna uma mensagem de erro se a divisão por zero for tentada.
        return "Erro: Divisão por zero não é permitida."

# --- Calculadora Interativa --- #

# Solicita ao usuário que insira o primeiro número.
num1_input = input("Digite o primeiro número: ")
# Converte a entrada do usuário para um número float.
num1 = float(num1_input)

# Solicita ao usuário que insira o segundo número.
num2_input = input("Digite o segundo número: ")
# Converte a entrada do usuário para um número float.
num2 = float(num2_input)

# Solicita ao usuário que escolha a operação.
operacao = input("Escolha a operação (+, -, *, /): ")

resultado = ""

# Realiza a operação baseada na escolha do usuário.
if operacao == '+':
    resultado = adicao(num1, num2)
elif operacao == '-':
    resultado = subtracao(num1, num2)
elif operacao == '*':
    resultado = multiplicacao(num1, num2)
elif operacao == '/':
    resultado = divisao(num1, num2)
else:
    resultado = "Erro: Operação inválida."

# Imprime o resultado da operação escolhida.
print(str(num1) + " " + operacao + " " + str(num2) + " = " + str(resultado))
```

# Jogo da Forma

```python
# Importa o módulo 'random' para permitir a escolha aleatória de elementos.
import random
# Importa o módulo 'unicodedata' para normalizar strings e remover acentos.
import unicodedata

# Dicionário que armazena as formas geométricas e suas descrições.
# A chave é o nome da forma (em minúsculas para facilitar a comparação) e o valor é uma frase descritiva.
formas_geometricas = {
    "quadrado": "Eu tenho 4 lados iguais e 4 ângulos retos.",
    "triângulo": "Eu tenho 3 lados e 3 vértices.",
    "círculo": "Eu não tenho lados, sou uma linha curva fechada.",
    "retângulo": "Eu tenho 4 lados, meus lados opostos são iguais e tenho 4 ângulos retos."
}

# Função para remover acentos de uma string.
# Isso permite que o usuário digite 'triangulo' e seja comparado corretamente com 'triângulo'.
def remover_acentos(texto):
    # Normaliza a string para uma forma decomposta (NFKD) para separar acentos dos caracteres base.
    nfkd_form = unicodedata.normalize('NFKD', texto)
    # Codifica para ASCII, ignorando caracteres que não podem ser representados (como os acentos),
    # e depois decodifica de volta para UTF-8.
    return nfkd_form.encode('ascii', 'ignore').decode('utf-8')

# --- Função Principal do Jogo --- #
# Esta função gerencia o fluxo do jogo de identificação de formas geométricas.
def jogar_forma_geometrica():
    # Imprime uma mensagem de boas-vindas e instruções para o jogador.
    print("Bem-vindo ao Jogo de Identificação de Formas Geométricas!")
    print("Você receberá uma descrição e deverá adivinhar qual é a forma.")
    print("Para sair do jogo a qualquer momento, digite 'sair'.")
    print("-" * 50) # Imprime uma linha divisória para melhorar a leitura.

    pontuacao = 0 # Inicializa a pontuação do jogador com zero.
    rodadas = 0   # Inicializa o contador de rodadas jogadas com zero.

    # Loop principal do jogo. Ele continuará executando novas rodadas
    # até que o jogador decida sair.
    while True:
        # Seleciona aleatoriamente o nome de uma forma geométrica do dicionário.
        # random.choice(list(dicionario.keys())) é usado para obter uma chave aleatória.
        forma_correta = random.choice(list(formas_geometricas.keys()))
        # Obtém a descrição correspondente à forma selecionada.
        descricao_forma = formas_geometricas[forma_correta]

        # Apresenta a descrição da forma ao usuário.
        print("\nDescrição: " + descricao_forma)
        # Solicita que o usuário digite sua resposta e a converte para minúsculas
        # para que a comparação com a forma_correta seja case-insensitive.
        resposta_usuario = input("Qual é a forma? ").lower()

        # Verifica se o usuário digitou 'sair' para encerrar o jogo.
        if resposta_usuario == 'sair':
            print("\n" + "-" * 50)
            # Imprime a pontuação final do jogador.
            print("Fim do jogo! Sua pontuação final é: " + str(pontuacao) + " de " + str(rodadas) + " rodadas.")
            print("Obrigado por jogar!")
            break # Sai do loop 'while True', terminando o jogo.

        rodadas += 1 # Incrementa o contador de rodadas a cada palpite.

        # Compara a resposta do usuário (após remover acentos) com a forma correta (também após remover acentos).
        if remover_acentos(resposta_usuario) == remover_acentos(forma_correta):
            pontuacao += 1 # Se a resposta estiver correta, a pontuação é incrementada.
            print("Correto! Muito bem!")
        else:
            # Se a resposta estiver incorreta, informa o erro e revela a forma correta.
            # .capitalize() é usado para que a primeira letra da forma seja maiúscula na saída.
            print("Incorreto. A forma correta era: " + forma_correta.capitalize())

        # Exibe a pontuação atual do jogador após cada rodada.
        print("Pontuação atual: " + str(pontuacao) + " de " + str(rodadas) + " rodadas.")
        print("-" * 50) # Imprime outra linha divisória.

# Chama a função principal para iniciar o jogo quando o script é executado.
jogar_forma_geometrica()
```

# API de Consulta de CEP

```python
import requests # Importa a biblioteca requests para fazer requisições HTTP a APIs externas.

def consultar_cep(cep):
    """
    Consulta a API pública ViaCEP para obter informações detalhadas de endereço a partir de um CEP.

    Args:
        cep (str): O Código de Endereçamento Postal a ser consultado. Pode conter formatação (ex: '01001-000').

    Returns:
        dict or None: Um dicionário contendo os dados do endereço se a consulta for bem-sucedida,
                      ou None se ocorrer algum erro (CEP inválido, não encontrado, problema de conexão, etc.).
    """
    # 1. Pré-processamento do CEP:
    # Remove quaisquer caracteres não numéricos (hífens, pontos, espaços) para garantir que apenas dígitos permaneçam.
    # Isso padroniza o CEP para o formato esperado pela API (8 dígitos numéricos).
    cep = str(cep).replace("-", "").replace(".", "").replace(" ", "")

    # 2. Validação básica do CEP:
    # Verifica se o CEP, após o pré-processamento, tem exatamente 8 dígitos e se todos são numéricos.
    # Se não atender a esses critérios, é considerado um CEP inválido e uma mensagem de erro é exibida.
    if len(cep) != 8 or not cep.isdigit():
        print("Erro: O CEP deve conter 8 dígitos numéricos válidos.")
        return None

    # 3. Construção da URL da API:
    # Monta a URL completa para a requisição à API ViaCEP, inserindo o CEP formatado.
    url = f"https://viacep.com.br/ws/{cep}/json/"

    # 4. Requisição à API e tratamento de erros:
    try:
        # Faz uma requisição GET para a URL da API.
        response = requests.get(url)
        # Verifica se a requisição foi bem-sucedida (status code 2xx).
        # Se for um status de erro (4xx ou 5xx), levanta um HTTPError.
        response.raise_for_status()

        # Tenta decodificar a resposta JSON da API.
        data = response.json()

        # 5. Verificação de erro na resposta da API:
        # A API ViaCEP retorna um campo 'erro': true se o CEP não for encontrado.
        if "erro" in data and data["erro"] == True:
            print(f"Erro: CEP '{cep}' não encontrado na base de dados do ViaCEP.")
            return None
        else:
            # Se não houver erro, retorna o dicionário com os dados do endereço.
            return data

    # 6. Tratamento de exceções específicas da biblioteca requests:
    except requests.exceptions.HTTPError as e:
        # Erro HTTP (ex: 404 Not Found, 500 Internal Server Error).
        print(f"Erro HTTP ao consultar o CEP {cep}: {e}. Verifique o formato do CEP e tente novamente.")
    except requests.exceptions.ConnectionError as e:
        # Erro de conexão (ex: sem internet, servidor ViaCEP fora do ar).
        print(f"Erro de conexão ao consultar o CEP {cep}: {e}. Verifique sua conexão com a internet ou tente mais tarde.")
    except requests.exceptions.Timeout as e:
        # Tempo limite excedido para a requisição.
        print(f"Tempo limite excedido ao consultar o CEP {cep}: {e}. A API demorou muito para responder.")
    except requests.exceptions.RequestException as e:
        # Captura qualquer outro erro inesperado da biblioteca requests.
        print(f"Ocorreu um erro inesperado na requisição para o CEP {cep}: {e}.")
    except ValueError:
        # Erro ao decodificar JSON (resposta da API não é um JSON válido).
        print(f"Erro: Resposta inválida da API para o CEP {cep}. Não foi possível decodificar o JSON.")

    # Em caso de qualquer erro tratado nas exceções, retorna None.
    return None

def exibir_endereco_formatado(dados_endereco):
    """
    Exibe as informações de endereço de forma legível e formatada.

    Args:
        dados_endereco (dict or None): Um dicionário contendo os dados do endereço,
                                       conforme retornado pela função `consultar_cep`.
    """
    # Verifica se os dados do endereço foram fornecidos (não são None).
    if dados_endereco:
        print("\n--- Endereço Encontrado ---") # Cabeçalho para a saída.
        # Usa .get() para acessar os valores do dicionário. Isso evita erros se uma chave não existir.
        print(f"CEP: {dados_endereco.get('cep', 'N/A')}")
        print(f"Logradouro: {dados_endereco.get('logradouro', 'N/A')}")
        print(f"Complemento: {dados_endereco.get('complemento', 'N/A')}")
        print(f"Bairro: {dados_endereco.get('bairro', 'N/A')}")
        print(f"Localidade: {dados_endereco.get('localidade', 'N/A')}")
        print(f"UF: {dados_endereco.get('uf', 'N/A')}")
        print(f"DDD: {dados_endereco.get('ddd', 'N/A')}")
        # Adicionando mais informações detalhadas, se disponíveis
        print(f"IBGE: {dados_endereco.get('ibge', 'N/A')}")
        print(f"GIA: {dados_endereco.get('gia', 'N/A')}")
        print(f"SIAFI: {dados_endereco.get('siafi', 'N/A')}")
        print("--------------------------") # Rodapé para a saída.
    else:
        # Mensagem exibida se não foi possível obter os dados do endereço.
        print("Não foi possível obter os dados do endereço para o CEP informado.")

# --- Uso interativo do script ---
# Este loop permite que o usuário digite múltiplos CEPs até que decida sair.
while True:
    # Solicita ao usuário que digite um CEP ou 'sair' para encerrar.
    cep_digitado = input("Digite o CEP para consulta (ou 'sair' para encerrar): ")

    # Converte a entrada para minúsculas e verifica se é 'sair'.
    if cep_digitado.lower() == 'sair':
        print("Consulta encerrada. Obrigado por usar o consultor de CEPs!")
        break # Sai do loop, encerrando o programa.

    # Chama a função para consultar o CEP digitado.
    dados = consultar_cep(cep_digitado)
    # Chama a função para exibir o endereço formatado com base nos dados obtidos.
    exibir_endereco_formatado(dados)
```
